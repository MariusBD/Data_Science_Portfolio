## EDA CATEGÓRICAS

### Análisis estadístico

```python
variables = ['antigüedad_empleo', 'ingresos_verificados', 'rating', 'vivienda',
       'finalidad', 'num_cuotas', 'estado']
```

```python
def frecuencias_cat(df_cat):
    resultado = df_cat[variables].apply(lambda x: x.value_counts(normalize = True)).T.stack()\
                .to_frame().reset_index()\
                .rename(columns={'level_0':'Variable','level_1':'Valor',0: "Frecuencia"})\
                .sort_values(by = ['Variable','Frecuencia'])
    return(resultado)
```

```python
frecuencias_cat(cat)
#  variable estado sera la variable target que vamos a utilizar para el modelo de PD(probability of default)
```

### Análisis gráfico

```python
def graficos_eda_categoricos(cat):
    
    #Calculamos el número de filas que necesitamos
    from math import ceil
    filas = ceil(cat.shape[1] / 2)

    #Definimos el gráfico
    f, ax = plt.subplots(nrows = filas, ncols = 2, figsize = (16, filas * 6))

    #Aplanamos para iterar por el gráfico como si fuera de 1 dimensión en lugar de 2
    ax = ax.flat 

    #Creamos el bucle que va añadiendo gráficos
    for cada, variable in enumerate(cat):
        cat[variable].value_counts().plot.barh(ax = ax[cada])
        ax[cada].set_title(variable, fontsize = 12, fontweight = "bold")
        ax[cada].tick_params(labelsize = 12)
```

```python
graficos_eda_categoricos(cat[variables])
```

## EDA NUMÉRICAS

### Análisis estadístico

```python
def estadisticos_cont(num):
    #Calculamos describe
    estadisticos = num.describe().T
    #Añadimos la mediana
    estadisticos['median'] = num.median()
    #Reordenamos para que la mediana esté al lado de la media
    estadisticos = estadisticos.iloc[:,[0,1,8,2,3,4,5,6,7]]
    #Lo devolvemos
    return(estadisticos)
```

```python
estadisticos_cont(num)
```

* dti ratio entre deuda y ingresos, no deberiamos tener -1 porque es %, deberiamos tener 0 y 100.
* porc_uso_revolving la parte superior deberia ser 100.
* minimo de prestamo que se esta concediendo es de 1000 y el maximo de 40k,lo cual encaja con los datos porque tenemos un dataset de prestamo personal

### Análisis gráfico

```python
def graficos_eda_continuas(num):
    
    #Calculamos el número de fila que necesitamos
    from math import ceil
    filas = ceil(num.shape[1] / 2)

    #Definimos el gráfico
    f, ax = plt.subplots(nrows = filas, ncols = 2, figsize = (16, filas * 6))

    #Aplanamos para iterar por el gráfico como si fuera de 1 dimensión en lugar de 2
    ax = ax.flat 

    #Creamos el bucle que va añadiendo gráficos
    for cada, variable in enumerate(num):
        num[variable].plot.density(ax = ax[cada])
        ax[cada].set_title(variable, fontsize = 12, fontweight = "bold")
        ax[cada].tick_params(labelsize = 12)
```

```python
graficos_eda_continuas(num)
```

* porc_tarjetas_75p: Hay bastante gente que ha hecho el consumo de todas las tarjetas de credito ya que estan al percentil 100.Podríamos discretizarlo en 3 grupos: 1. los que no tienen nada de uso en las tarjetas,2.los que si tienen uso, 3. los que han usado el 100%

* principal: la gran mayoria sin importes bajos, por debajo de los 10k y va decayendo hasta llegar a los 40k

* tipo de interes: parece que tenemos un tp bastante medio-alto, entorno al 12-15%.

* imp_recuperado:parece ser que los prestamos que han resultado ser impagados, no se recuperan.

Vamos a analizar el detalle de imp_amortizado e imp_recuperado, ya que son relevantes para la target del modelo LGD.

```python
ratios = num[['principal','imp_amortizado','imp_recuperado']].copy()
```

```python
#  El r_amortizado_principal nos va decir el % del importe prestamo inicial que el cliente ya pagado
ratios['r_amortizado_principal'] = ratios.imp_amortizado / ratios.principal
ratios['r_recuperado_principal'] = ratios.imp_recuperado / ratios.principal
```

```python
ratios.describe().T
'''el hecho de que hemos recuperado mas que el prestamo inicial, puede significar algo, que alomejor al cliente se le ha 
aplicado algo extra'''
```

Conclusiones:

* En transformación convertir num_derogatorios a binaria
* Revisar en profundidad la relación entre estado e imp_recuperado, ya que habia una serie de valores que no sabiamos si asignarlos a impagos o no.Solo se puede tener imp_recuperado en aquellos casos en los cuales haya habido impagos.Hay que cruzar estas dos variables, para ver que valores de la variable estado si que estamos comprpobando que tendríamos que asignar a la categoria de impago.Si hay valor recuperado tendriamos impago.

```python
temp = cat[['estado']].merge(num[['imp_recuperado']], left_index=True, right_index=True)
temp.groupby('estado').imp_recuperado.mean()
```

Efectivamente sólo hay importe recuperado en el estado Charged Off.

Las categorias estado y Does not meet the credit policy. Status:Charged Off, son impagos porque tienen importes recuperados.Las otras categorias no significa que no sean impago,podrían ser impago pero no se ha podido recuperar nada.

Esto sirve para la variable target del modelo PD, sabemos que dos valores que si o si tienen que estar metidos dentro de la categoria de impagados.Charged Off y Does not meet the credit policy. Status:Charged Off
