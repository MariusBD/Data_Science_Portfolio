Vamos a reescalar ya que el algoritmo que vamos a usar es el K-medias, este algoritmo era sensible a diferentes escalas. De los tres metodos el Robust Scaler lo descartamos ya que no tenemos problemas de atipicos.

Min-Max-Scaling es el que vamos a usar, ya que la mayoria de nuestras variables categoricas,les vamos a aplicar OHE, eso significa que las vamos a transformar en un rango de 0 a 1 y el MMS nos permite tambien llevar estas variables tambien en un rango de 0-1.

En este caso concreto como vamos a hacer un cluster todavia es mas importante aplicar MMS junto con OHE.Si hicieramos modelizacion predictiva no pasaria nada por tener todas las variables categoricas con el rango 0-1 y tener las numericas SS en un rango de -4 y +4.

## TRANSFORMACIÓN DE CATEGÓRICAS

### One Hot Encoding

#### Variables a aplicar OHE

```python
var_ohe = ['origen','fuente','ult_actividad','ambito','ocupacion','descarga_lm']
```

#### Instanciar

```python
ohe = OneHotEncoder(sparse = False, handle_unknown='ignore')
```

#### Entrenar y aplicar

```python
cat_ohe = ohe.fit_transform(cat[var_ohe])
```

#### Guardar como dataframe

```python
cat_ohe = pd.DataFrame(cat_ohe, columns = ohe.get_feature_names_out())
```

## TRANSFORMACIÓN DE NUMÉRICAS

NO TENEMOS QUE TRANSFORMAR NINGUNA VARIABLE NUMERICA.SOLO VAMOS A REESCALAR.

## UNIFICAR DATASETS TRANSFORMADOS

```python
cat_ohe
```

```python
num
```

```python
df = pd.concat([cat_ohe,num.reset_index()], axis = 1)
df
```

## REESCALAR VARIABLES

### Con Min-Max

```python
df.info()
```

#### Variables a reescalar con Min-Max

```python
var_mms = df.iloc[:,39:].columns
```

#### Instanciar

```python
mms = MinMaxScaler()
```

#### Entrenar y aplicar

```python
df_mms = mms.fit_transform(df[var_mms])
```

#### Guardar como dataframe

```python
#Añadir sufijos a los nombres
nombres_mms = [variable + '_mms' for variable in var_mms]

#Guardar como dataframe
df_mms = pd.DataFrame(df_mms,columns = nombres_mms)
```

## UNIFICAR DATASETS REESCALADOS

### Crear una lista con los dataframes a incluir en el tablón analítico
